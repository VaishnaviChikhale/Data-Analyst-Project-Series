**About Iris Dataset:**

* The Iris dataset, introduced by the statistician and biologist Ronald A. Fisher in 1936, is a classic dataset used for machine learning and statistics.

*  There are 6 rows in dataset as Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, and Species. There are 150 rows or entries.

* It contains 150 observations of iris flowers, with each sample having four features: sepal length, sepal width, petal length, and petal width.

* The dataset is divided into three species of iris flowers: Setosa, Versicolor, and Virginica, with 50 samples from each species.

* It is commonly used for testing classification algorithms and for educational purposes to understand data analysis and pattern recognition.

* The data is typically presented in a tabular format, where each row represents an iris sample, and each column represents a feature or the class label.

```python
colors= ['red', 'orange', 'blue']
species= ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']
```

```python
for i in range(3):
  x= df[df['Species']== species[i]]
  plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'],c= colors[i], label= species[i])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
```

```python
for i in range(3):
  x= df[df['Species']== species[i]]
  plt.scatter(x['PetalLengthCm'], x['PetalWidthCm'], c= colors[i], label= species[i])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
```

```python
for i in range(3):
  x= df[df['Species']== species[i]]
  plt.scatter(x['SepalLengthCm'], x['PetalLengthCm'], c= colors[i], label= species[i])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend
```

```python
for i in range(3):
  x= df[df['Species']== species[i]]
  plt.scatter(x['SepalWidthCm'], x['PetalWidthCm'], c= colors[i], label= species[i])
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.legend
```

**Insights:**

Iris-setosa stands out from Versicolor and Virginica making it easy to spot. When you look at petal length versus petal width, the classes don't mix showing clear differences. But when it comes to sepal size, Versicolor and Virginica overlap a lot, which means it's harder to tell them apart.

```python
numeric_df = df.select_dtypes(include=[float, int])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
sns.heatmap(correlation_matrix , annot=True, ax=ax, cmap = 'mako')
```

**Insights:**

Petal length and petal width are strongly positively correlated, with a correlation coefficient of 0.96. This means that as petal length increases, petal width tends to increase as well. Additionally, sepal length has a strong positive correlation with both petal length and petal width, suggesting that as sepal length grows, so do the petal measurements. On the other hand, sepal width shows a negative correlation with petal length and petal width, indicating that as sepal width increases, both petal length and width tend to decrease.
